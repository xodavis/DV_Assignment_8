```python
import pandas as pd
import altair as alt

df = pd.read_csv ("emissions.csv")
df1 = df[1:4997]
df2 = df[4998:8386]
```

```python
nations = df2[df2["Country"].isin(["United States", "Russia", "United Kingdom", "Yemen", "Spain and Andorra"])]

highlight = alt.selection(type = "single", on = "mouseover", fields = ["Country"], nearest = True)

nat = alt.Chart(nations).mark_line().encode(
    alt.X("Year:Q",scale = alt.Scale(domain = (1970, 2012))),
    alt.Y("Emissions_Type_CO2:Q", title = "CO2 Emissions"),
    color=alt.condition(highlight, "Country:N", alt.value("lightgray"))
)

points = nat.mark_circle().encode(
    opacity=alt.value(0)
).add_selection(
    highlight
)

points + nat

#link i referenced from: https://altair-viz.github.io/gallery/multiline_highlight.html
```

```python
countries = df1[df1["Country"].isin(["Brazil", "Colombia", "France and Monaco", "Congo", "Egypt"])]

select = alt.selection(type = "single", on='mouseover', fields = ["Country"], bind = "legend", nearest = True)

count = alt.Chart(countries).mark_line().encode(
    alt.X("Year:Q",scale = alt.Scale(domain = (1970, 2012))),
    alt.Y("Emissions_Type_N2O:Q", title = "NO2 Emissions"),
    color=alt.condition(select, "Country:N", alt.value("lightgray"))
).add_selection(select)

count
```

```python
nigeria = df2[df2["Country"] == "Nigeria"]

alt.Chart(nigeria, title = "Nigeria Transportation Emissions").mark_trail().encode(
    alt.X("Year:Q", scale = alt.Scale(domain = (1970, 2012))),
    alt.Y("Emissions_Sector_Transport", title = "Transportation Emissions"),
    size = "Emissions_Sector_Transport"
).interactive()
```
